Create a "Fish" class with the properties length, brightness, and a text label describing which class (i.e., Salmon/Sea Bass) it belongs to.

```python
class Fish(object):
  def __init__(self,length,brightness,label):
    self.length=length
    self.brightness = brightness
    self.label = label

  def distanceTo(self,r):
    #TODO: the function should return the Euclidian distance between the object (i.e., "self") and a reference object r
    return(math.sqrt((self.length-r.length)**2 + (self.brightness-r.brightness)**2))

print("class fish interpreted")
```

```python
import random
class KMeans(object):
  def __init__(self,array):
    self.array = array

  def cluster(self,n):
    if((n<=1) or (n>len(self.array))):
      return "more than 1 cluster needed"

    #TODO implement the k-Means clustering by creating a list of cluster centers (clusters[]) and a list of items that belong to each cluster center (clusterItems[]), where the index corresponds between both array

    # Placing n centroids at random locations (choosing n random fish as reference centroids)
    clusters = random.sample(self.array, n)
    clusterItems = []
    for i in range(n):
      clusterItems.append([])

    #print(clusters)

    isFinished = False

    while not isFinished:
            # "clear" clusterItems array
            for i in range(n):
                clusterItems[i] = []

            # find nearest centroid for each fish
            for fish in self.array:
              distance = [] # array of distances per fish
              for j in range(n):
                distance.append(fish.distanceTo(clusters[j]))
              nearest_index = distance.index(min(distance))
              clusterItems[nearest_index].append(fish)

            # calculate clusters for current run
            temp_clusters = []
            for i in range(n):

                avg_length = sum(f.length for f in clusterItems[i]) / len(clusterItems[i])
                avg_brightness = sum(f.brightness for f in clusterItems[i]) / len(clusterItems[i])
                temp_clusters.append(Fish(avg_length, avg_brightness, ""))

            # if cluster centers stayed the same as in the last run -> stop
            for j in range(n):
                if temp_clusters[j].length == clusters[j].length and temp_clusters[j].brightness == clusters[j].brightness:
                  isFinished = True

            clusters = temp_clusters

    print("\n")
    for i in range(n):
      print("Cluster " +str(i)+" "+str(clusters[i].length)+" "+str(clusters[i].brightness))
      for u in range(len(clusterItems[i])):
        print(clusterItems[i][u].label+" "+str(clusterItems[i][u].length)+" "+str(clusterItems[i][u].brightness))


```
